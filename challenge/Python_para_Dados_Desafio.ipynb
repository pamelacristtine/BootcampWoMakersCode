```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python

saude_estilo_vida = 'saude_do_sono_estilo_vida.csv'

df_estilo_vida = pd.read_csv(saude_estilo_vida)

display(df_estilo_vida.head())
```

**Questão 1**

```python
df_estilo_vida.columns
```

```python
# Renomear colunas do dataset
rename_saude = {
    'ID': 'Identificador',
    'Pressão sanguíneaaaa': 'Pressão Sanguínea',
    'Ocupação': 'Profissão',
    'Categoria BMI' :'Categoria IMC'
}

df_estilo_vida = df_estilo_vida.rename(columns= rename_saude)
display(df_estilo_vida.head())

```

**Questão 2**

```python
#Estatística por Profissão (Média | Moda | Mediana)

resultado_estatistica = df_estilo_vida.groupby('Profissão')['Duração do sono'].agg(['mean', 'median', lambda x: x.mode()[0]])

resultado_estatistica.columns = ['media', 'mediana','moda']

print(resultado_estatistica)

```

**Questão 3**

```python
#Obesidade em pessoas que atuam como Eng. de Software

filtro_obesos_engsoft= df_estilo_vida.loc[(df_estilo_vida['Profissão'] == 'Eng. de Software') & (df_estilo_vida['Categoria IMC'] == 'Obesidade')]

filtro_engsoft= df_estilo_vida.loc[(df_estilo_vida['Profissão'] == 'Eng. de Software')]

obesos_engsoft = filtro_obesos_engsoft.groupby('Profissão')['Categoria IMC'].count()
engsoft = filtro_engsoft.groupby('Profissão') ['Categoria IMC'].count()

porc_obesos_engsoft = (obesos_engsoft/engsoft) * 100

print(f"A porcentagem de obesos que atuam com Eng. de Software é:\n{porc_obesos_engsoft}%")


```

**Questão 4**

```python
valores_distintos = df_estilo_vida['Profissão'].unique()
print(valores_distintos)
```

```python
#Advogar ou ser representante de vendas faz você dormir menos
profissao= ['Representante de Vendas', 'Advogado(a)']
df_filtrado = df_estilo_vida[df_estilo_vida['Profissão'].isin(profissao)]
media_sono = df_filtrado.groupby('Profissão')['Duração do sono'].mean()

prof_menos_horas_sono= media_sono.idxmin()
horas_sono= round(media_sono[prof_menos_horas_sono], 2)

print(f"De acordo com os dados, ser {prof_menos_horas_sono} faz você dormir menos. Com uma média de {horas_sono} horas de sono por dia.")
```

**Questão 5**

```python
#Entre quem fez enfermagem e quem fez medicina quem tem menos horas de sono

profissao= ['Médico(a)', 'Enfermeiro(a)']
df_filtrado = df_estilo_vida[df_estilo_vida['Profissão'].isin(profissao)]
media_sono = df_filtrado.groupby('Profissão')['Duração do sono'].mean()

prof_menos_horas_sono= media_sono.idxmin()
horas_sono= round(media_sono[prof_menos_horas_sono], 2)

print(f"De acordo com os dados, ser {prof_menos_horas_sono} faz você ter menos horas de sono. Com uma média de {horas_sono} horas de sono por dia.")

```

**Questão 7**

```python
#Profissão menos frequente no conjunto

prof_freq = df_estilo_vida['Profissão'].value_counts()
prof_menos_freq= prof_freq.idxmin()
print(f"De acordo com os dados, a profissão menos frequente é a de {prof_menos_freq}!")
```

**Questão 8**

```python
#Quem tem maior pressão sanguínea média(homens ou mulheres?)

df_estilo_vida[['Pressão Sistólica', 'Pressão Diastólica']] = df_estilo_vida['Pressão Sanguínea'].str.split('/', expand=True).astype(int)

media_pressao_genero = df_estilo_vida.groupby('Gênero')['Pressão Sistólica', 'Pressão Diastólica'].mean()

maior_media_pressao_genero= media_pressao_genero.idxmax()

print(f"O gênero que possui maior Pressão Sistólica e Pressão Diastólica é: \n{maior_media_pressao_genero}")
```

**Questão 9**

```python
#É predominante entre os participantes dormir 8 horas por dia

moda = df_estilo_vida['Duração do sono'].mode()[0]

if moda == 8:
  print('É predominante entre os participantes dormir 8 horas por dia!')

else:
  print('Não é predominante entre os participantes dormir 8 horas por dia!')
```

**Questão 10**

```python
#Frequência cardíaca e Passos diários

freq_card_maior70= df_estilo_vida.loc[(df_estilo_vida['Frequência cardíaca'] >= 70)]

freq_card_menor70= df_estilo_vida.loc[(df_estilo_vida['Frequência cardíaca'] < 70)]


media_maior70 = round(freq_card_maior70['Passos diários'].mean(),2)

media_menor70 = round(freq_card_menor70['Passos diários'].mean(),2)

if media_maior70 > media_menor70:
  print('Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70!')

else:
  print('Pessoas com frequências cardíacas abaixo de 70 dão mais passos que pessoas com frequência cardíaca maior ou igual a 70')

```
