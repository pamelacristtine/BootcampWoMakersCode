```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

```python
#Importando o dataset
enem_2023 = 'enem_2023.json'

# Leitura do arquivo Json
df_enem_2023 = pd.read_json(enem_2023)

#Imprimir na tela o dataset
display(df_enem_2023.head())
```

**Questão 1**

```python
materias= df_enem_2023[['Linguagens','Ciências humanas','Ciências da natureza','Matemática','Redação']]

amplitude = materias.max() - materias.min()

max_amplitude = round(amplitude.max(),2)
nome_coluna_max_amplitude= amplitude.idxmax()
print(f"A coluna com a máxima amplitude é: {nome_coluna_max_amplitude}\nCom uma amplitude de: {max_amplitude}")
```

**Questão 2**

```python
print('Média para cada uma das disciplinas:\n')
media_disciplina = round(df_enem_2023.mean(),2)
print(media_disciplina)

print('\n')
print('Mediana para cada uma das disciplinas:\n')

mediana_disciplina =round(df_enem_2023.median(skipna=True), 2)
print(mediana_disciplina)
```

**Questão 3**

```python
df_enem_peso = df_enem_2023
df_enem_peso['LxP'] = df_enem_peso['Linguagens'] * 2
df_enem_peso['CHxP'] = df_enem_peso['Ciências humanas'] * 1
df_enem_peso['CNxP'] = df_enem_peso['Ciências da natureza'] * 1
df_enem_peso['MxP'] = df_enem_peso['Matemática'] * 4
df_enem_peso['RxP'] = df_enem_peso['Redação'] * 2

df_enem = df_enem_peso.loc[:, ['LxP','CHxP','CNxP','MxP','RxP', 'Sexo']]
df_enem_500 = df_enem.sort_values(by=['LxP','CHxP','CNxP','MxP','RxP'], ascending=False).head(500)

media = df_enem_500.mean()
desvio_padrao = np.std(df_enem_500)

print('Média:\n')
print(media)

print('\n')
print('Desvio padrão:\n')
print(desvio_padrao)

```

**Questão 4**

```python
df_enem_40 = df_enem.sort_values(by=['LxP','CHxP','CNxP','MxP','RxP'], ascending=False).head(40)

media_ccomp = np.mean(df_enem_40)
variancia_ccomp = np.var(df_enem_40)

print('Média da nota dos estudantes que entraram no curso de ciência da computaçã:\n')
print(media_ccomp)

print('\n')
print('Variância da nota dos estudantes que entraram no curso de ciência da computação:\n')
print(variancia_ccomp)

```

**Questão 5**

```python
#removendo nulos
mat_sem_nulos= df_enem_2023['Matemática'].dropna()

ling_sem_nulos= df_enem_2023['Linguagens'].dropna()

#quartil75

terceiro_quartil_mat = round(np.percentile(mat_sem_nulos, 75),2)

terceiro_quartil_ling = round(np.percentile(ling_sem_nulos, 75),2)

print(f"Terceiro quartil para a disciplina de Matemática: {terceiro_quartil_mat} \n")
print(f"Terceiro quartil para a disciplina de Linguagens: {terceiro_quartil_ling}")
```

**Questão 6**

```python

notas_redacao = df_enem_2023['Redação'].dropna()
notas_linguagens = df_enem_2023['Linguagens'].dropna()

intervalos = range(0,1000, 20)

# Histograma Redação
plt.figure(figsize=(20, 10))
plt.hist(notas_redacao, bins=intervalos, edgecolor='black', alpha=0.5, label='Redação')
plt.title('Histograma de Notas Redação')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.xticks(intervalos)
plt.show()


# Histograma Linguagens
plt.figure(figsize=(20, 10))
plt.hist(notas_linguagens, bins=intervalos, edgecolor='black', alpha=0.5, label='Linguagens')
plt.title('Histograma de Notas Linguagens')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.xticks(intervalos)
plt.show()

```

R.:

Podemos observar que os histogramas são simétricos, pois, a distribuição dos dados é simétrica em relação a um ponto central.
